```python
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import numpy as np
```

```python
certificates = torch.load('output/cora_ml/certificates')
noise = list(certificates.keys())[0]
certificates = certificates[noise]

certificate_grid = certificates['certificate_grid']
correctly_classified = certificates['correctly_classified']
abstain = certificates['abstain']

print('Test set accuracy', np.mean(correctly_classified))
print('Number of abstains', abstain.sum())

overall = np.zeros_like(certificate_grid[0])
for i in range(len(certificate_grid)):
    if correctly_classified[i] and not abstain[i]:
        overall += (certificate_grid[i] > 0.5)
overall = overall / len(certificate_grid)
overall

fig, ax = plt.subplots(figsize=(15, 6))
sns.heatmap(overall[:5, :7], square=True, vmin=0.0, vmax=1.0, annot=True, linewidths=0.1, ax=ax)
ax.invert_yaxis()
ax.set_xlabel('Edges added.')
ax.set_ylabel('Edges deleted.')
```

```python
def plot_certificate(certificate, ax):
    certificate_grid = certificate['certificate_grid']
    correctly_classified = certificate['correctly_classified']

    overall = np.zeros_like(certificate_grid[0])
    for i in range(len(certificate_grid)):
        if correctly_classified[i] and not abstain[i]:
            overall += (certificate_grid[i] > 0.5)
    overall = overall / len(certificate_grid)
    overall

    sns.heatmap(overall, square=True, vmin=0.0, vmax=1.0, annot=True, linewidths=0.01, cbar=False, ax=ax)
    ax.invert_yaxis()

certificates = torch.load('output/synthetic/certificates')
noise = list(certificates.keys())
xs = np.unique(np.array(noise)[:, 0])
ys = np.unique(np.array(noise)[:, 0])

fig, axs = plt.subplots(len(xs), len(ys), figsize=(25, 25))
for i, x in enumerate(xs):
    for j, y in enumerate(ys):
        ax = axs[-(i+1)][j]
        plot_certificate(certificates[(x, y)], axs[i, j])
        ax.set_title(fr'$p_i={round(x,2)}$ $p_o={round(y, 2)}$')
        if j == 0:
            axs[i, j].set_ylabel('Edges between community')
        if i == len(xs) - 1:
            axs[i, j].set_xlabel('Edges in community')
```
